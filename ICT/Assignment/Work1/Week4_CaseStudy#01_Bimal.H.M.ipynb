### Q3 : Bar graph of Male v/s Female Buyers

```python
plt.title("Bar Graph of Male v/s Female buyers ")
CP = sns.countplot(x = 'Buyer Gender', data = cars)
for i in CP.patches:
    CP.annotate(i.get_height() , (i.get_x()+0.3, i.get_height()-2000))   #To Display count in graph
plt.show()
```

### Q4 : Top 5 Cars Based on their Sales Price

#### A) Top 5 Based on direct sorting of Sale Price

```python
sort1 = cars.nlargest(5,['Sale Price'])                      # In Sort1 Storing 5 Largest columns of dataset with respect to Sale Price
set1 = sort1[['Make','Model','Sale Price']].reset_index()    # In Set1 storing 3 Columns(Make,Model,Sale Price) from Sorted dataset by indexing from 0 to 4  
set1.index = np.arange(1,len(set1)+1)                        # Changing dataframe index of Set1 0 -> 4 to 1 -> 5 and print along with index from original data 
set1
```

#### B) Top 5 Different Makes based on Sale Price

```python
sort2 = cars.nlargest(15,['Sale Price'])           # In sort2 Storing 15 Largest columns of dataset with respect to Sale Price         
set2 = sort2[['Make','Model','Sale Price']]        # In set2 storing Make, Model, Sale Price columns from sorted dataset(Sort2)                     
set2 = set2.drop_duplicates(subset = ['Make'])     # Removing duplicate value with respect to Meke column
set2 = set2.head(5).reset_index()                  # Storing top 5 rows from set2 in itself and reseting indexing from 0 to 4  
set2.index = np.arange(1,len(set2)+1)              # Changing dataframe index of Set1 0 -> 4 to 1 -> 5 and print along with index from original data
set2
```

### Q5 : Least 5 Cars based on Resell Price

```python
sort3 = cars.nsmallest(5,['Resell Price'])[['Make','Model','Resell Price']]  # In sort3 storing 5 smallest values of Resell Price along with its make and model
sort3 = sort3.reset_index()                                                  # Assigning new indexing from 0 to 4
sort3.index = np.arange(1,len(sort3)+1)                                      # Changing indexing value to 0 to 5
sort3
```
